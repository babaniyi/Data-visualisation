```python
import matplotlib.pyplot as plt
import numpy as np
from drawarrow import ax_arrow
from highlight_text import ax_text
import matplotlib.patches as patches
```

```python

# Generate sample data
np.random.seed(42)
values = np.random.rand(15)+0.4  # 15 values
errors = np.random.rand(15) * 0.4  # Error margins
labels = [f'Item {i+1}' for i in range(15)]
labels = ['Liora', 'Zane', 'Mila', 'Taryn', 'Nico', 'Elora', 'Juno', 'Kai', 'Soren', 'Rhea', 'Mavis', 'Cleo', 'Orion', 'Ezra', 'Lina']

# Sort values in ascending order (opposite of descending)
sorted_indices = np.argsort(values)
values = 66*values[sorted_indices]
errors = 66*errors[sorted_indices]
labels = [labels[i] for i in sorted_indices]

# Create the horizontal lollipop plot
fig, ax = plt.subplots(figsize=(4, 3.5), dpi=120)
y_positions = np.arange(len(values))

# Plot the stems
ax.hlines(y_positions, xmin=0, xmax=values, color='#A4BED5FF', zorder=2)
# Plot the lollipop heads
ax.plot(values, y_positions, 'o', color='white', zorder=3, markersize=9)
ax.plot(values, y_positions, 'o', color='#476F84FF', zorder=3)
# Add error bars
ax.hlines(y_positions, xmin=values-errors, xmax=values+errors, color='#A4BED5FF', zorder=2, linewidth=6)

# Set y-axis labels and positions
ax.set_yticks(y_positions)
ax.set_yticklabels(labels)
ax.invert_yaxis()  # Highest index at the top

# Add spacing between lollipops
ax.set_ylim(-0.5, len(values) - 0.5)

# Remove the frame
for spine in ax.spines.values():
    spine.set_visible(False)

# Remove ticks
ax.tick_params(left=False, bottom=False)

# Add vertical grid lines
ax.xaxis.grid(True, linestyle='--', color='gray', alpha=0.3)

ax_arrow(
   tail_position=[77,3],
   head_position=[67,6.8],
   width=0.5,
   color = 'black',
   double_headed=False,
   fill_head=False,
   head_width=2,
   head_length=6,
   radius=0.2,
   ax=ax,
)

rect = patches.Rectangle(
    (75,1),  # (x, y) bottom-left corner
    20,  # Width
    2,  # Height
    facecolor="white",  # White fill
    edgecolor="white",  # No border
    zorder = 2
)
ax.add_patch(rect) 

ax_text(
    x=63, y=1.3,
    s="Focus on what\nmatters",
    fontsize=8,
    va='bottom',
    ha='left'
)

plt.tight_layout()
#plt.savefig("lollipop.png", dpi=300)
plt.show()
```
