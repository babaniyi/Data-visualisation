```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

url = "https://docs.google.com/spreadsheets/d/1kZ82O6oVcIHBfwYHG_yZRVc_cWQKI_xtyAhORc_DajE/export?format=csv&id=1kZ82O6oVcIHBfwYHG_yZRVc_cWQKI_xtyAhORc_DajE&gid=0"

#response = open_url(url)
df = pd.read_csv(url)

x = df['Tournament']
y = df['Cap']

fig, ax = plt.subplots(figsize=(12,12))

sns.swarmplot(x=x,y=y, ax=ax, 
              color = 'lightgrey', s = 9, linewidth=0)

sns.violinplot(x=x,y=y,hue=x, ax=ax, palette = 'Set2',
              fill = True,
              inner = None,
              width = 0.5)

ax.spines[:].set_visible(False)

ax.tick_params(length = 0, labelsize = 14, pad = -20)
ax.tick_params(axis='x', colors='black')
ax.set_ylabel('Stadium Capacity', color = 'grey', fontsize = 14)
ax.set_xlabel('')
ax.set_yticks(np.arange(40000, 105001, 15000),
             labels = ['40k','55k','70k','85k','100k'],
             color = 'grey',
             fontsize = 14)

fig.text(x = 0.095, y=0.85, 
         s = "Bigger Arenas, Bigger Doubts: 2026 WC faces a new scale", 
         fontsize=18, fontweight = 'bold')

fig.text(x = 0.095, y=0.785, s = """A visual breakdown of stadium capacity distributions across recent World Cups
adds to a growing concern: Can the U.S., despite visa issues and lukewarm football
culture fill so many large stadiums?""", 
         fontsize=14, color = 'grey')


plt.show()
```
